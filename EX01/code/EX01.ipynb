```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Scenario 1

```python
df = pd.read_csv('../data/data.csv')
print(df)
print("Head")
print(df.head())
print("Tail")
print(df.tail())        
print("info")
print(df.info())
print("describe")
print(df.describe())        
```

```python
df['Sales'] = df['Quantity'] * df['UnitPrice']
sales_per_product = df.groupby('StockCode')['Sales'].sum().head(20)

plt.bar(sales_per_product.index, sales_per_product.values)
plt.xlabel('Product')
plt.ylabel('Sales')
plt.title('Sales per Product')
plt.xticks(rotation=45)
plt.show()
```

# Scenario 3

```python
df=pd.read_csv('../data/Housing.csv')
print(df.head())
df.info()
df.notnull().sum()
```

```python
sns.scatterplot(x='area', y='price', data=df, hue="area")
plt.title('Price vs Area')
plt.show()
```

```python
sns.scatterplot(x='bedrooms', y='price', data=df, hue="furnishingstatus")
plt.title('Price vs BHK')
plt.show()
```

```python
num_df = df[['price', 'area', 'bedrooms', 'bathrooms', 'stories', 'parking']]
corr = num_df.corr()

sns.heatmap(corr, annot=True)
plt.title('Correlation Heatmap')
plt.show()
```

# Scenario 4

```python
df=pd.read_csv('../data/marketing_campaign.csv')
df.info()
df.isnull().sum()
```

```python
df['Age'] = 2024 - df['Year_Birth']
df['Total_Spending'] = (
    df['MntWines'] + df['MntFruits'] + df['MntMeatProducts'] +
    df['MntFishProducts'] + df['MntSweetProducts'] + df['MntGoldProds']
)
```

```python
sns.barplot(x='Education', y='Total_Spending', data=df, hue='Education', ci=None)
plt.title('Average Spending by Education Level')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.boxplot(y='Income', data=df)
plt.title('Income Distribution')
plt.show()
```

```python
sns.boxplot(x='Response', y='Age', data=df)
plt.title('Age Distribution by Campaign Response')
plt.show()
```
